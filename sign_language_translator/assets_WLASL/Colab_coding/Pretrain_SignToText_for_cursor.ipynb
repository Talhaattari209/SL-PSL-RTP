````python
import sign_language_translator as slt
import mediapipe as mp
import cv2
import numpy as np
import os
import pickle
from google.colab import drive

mp_holistic = mp.solutions.holistic.Holistic(
    min_detection_confidence=0.1,
    min_tracking_confidence=0.1,
    static_image_mode=True
)

def extract_landmarks_from_video(video_path, output_dir):
    video_name = os.path.splitext(os.path.basename(video_path))[0]
    output_file = os.path.join(output_dir, video_name, 'landmarks_wlasl_complete.pkl')
    if os.path.exists(output_file):
        os.remove(output_file)
        print(f"Overwriting existing {output_file}")

    cap = cv2.VideoCapture(video_path)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    frames = []
    for i in range(frame_count):
        ret, frame = cap.read()
        if not ret:
            break
        frames.append(frame)
        cv2.imwrite(f"/content/frame_{i}.jpg", frame)
    cap.release()
    print(f"Extracted {len(frames)} frames")

    landmarks = []
    for frame_idx in range(len(frames)):
        frame = frames[frame_idx]
        frame = cv2.resize(frame, (640, 480))
        results = mp_holistic.process(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        frame_landmarks = np.zeros((543, 3))
        if results.pose_landmarks:
            pose = np.array([[lm.x, lm.y, lm.z] for lm in results.pose_landmarks.landmark])
            frame_landmarks[0:33, :] = pose
        if results.face_landmarks:
            face = np.array([[lm.x, lm.y, lm.z] for lm in results.face_landmarks.landmark])
            frame_landmarks[33:501, :] = face
        if results.right_hand_landmarks:
            right_hand = np.array([[lm.x, lm.y, lm.z] for lm in results.right_hand_landmarks.landmark])
            frame_landmarks[501:522, :] = right_hand
            print(f"Right hand detected in frame {frame_idx}")
        if results.left_hand_landmarks:
            left_hand = np.array([[lm.x, lm.y, lm.z] for lm in results.left_hand_landmarks.landmark])
            frame_landmarks[522:543, :] = left_hand
            print(f"Left hand detected in frame {frame_idx}")
        landmarks.append(frame_landmarks)
    landmarks = np.array(landmarks)
    os.makedirs(os.path.dirname(output_file), exist_ok=True)
    with open(output_file, 'wb') as f:
        pickle.dump(landmarks, f)
    print(f"Saved {video_name}: Shape {landmarks.shape}, Hands: {np.any(landmarks[:, 501:543, :] != 0)}")

video_path = '/content/drive/MyDrive/Datasets_ISL-CSLRT-PSL/Common_WLASL_videos'
output_dir = '/content/Landmarks_WLASL'
extract_landmarks_from_video(video_path, output_dir)
mp_holistic.close()
````

# Pre proccessing

### Key Features
1. **Input Handling**:
   - Loads existing `landmarks_complete.pkl` files from subdirectories in `'/content/drive/MyDrive/landmarks_data_corrected'`.
   - Skips files already preprocessed (if `landmarks_preprocessed.pkl` exists) to avoid redundant processing.
   - Includes error handling for corrupted or unreadable `.pkl` files.

2. **Preprocessing Assessment** (`assess_preprocessing_need`):
   - **Frame Count**: Checks if the number of frames is close to 45 (within 20% tolerance). Fewer frames indicate padding is needed.
   - **Landmark Variability**: Computes variance of x, y coordinates for hand landmarks (501:543). High variance (>0.1) suggests normalization and smoothing are needed to reduce noise.
   - **Missing Landmarks**: Calculates the proportion of frames with zeroed hand landmarks. A high ratio (>50%) indicates preprocessing is needed to handle missing data.
   - Outputs an analysis dictionary and a boolean (`preprocessing_needed`) to decide whether to apply preprocessing.

3. **Preprocessing Steps**:
   - **Normalization** (`normalize_landmarks`): Normalizes x, y coordinates separately for right (501:522) and left (522:543) hands, preserving gesture-specific dynamics. Skips zeroed frames to avoid distorting statistics.
   - **Smoothing** (`smooth_landmarks`): Applies a Gaussian filter (sigma=1) to non-zero frames, reducing jitter in landmark trajectories.
   - **Padding** (`pad_landmarks`): Replicates the last valid frame to reach 45 frames, avoiding noise from zero-padding. Truncates if more than 45 frames.

4. **Output**:
   - Saves preprocessed landmarks as `landmarks_preprocessed.pkl` in the same subdirectory as the input file.
   - Prints analysis results and preprocessing decisions for each video. Example output:
     ```
     Loaded landmarks for pk-hfad-1_1: Shape (50, 543, 3)
     Preprocessing analysis for pk-hfad-1_1: {'frame_count': 50, 'frame_count_sufficient': True, 'hand_variance': 0.12, 'high_variability': True, 'missing_hand_ratio': 0.2, 'excessive_missing': False}
     Preprocessing needed
     Saved preprocessed landmarks for pk-hfad-1_1: Shape (45, 543, 3), Hands: True
     ```

### Why Preprocessing May Be Needed
The `assess_preprocessing_need` function checks for:
- **Inconsistent Frame Counts**: If videos have significantly fewer than 45 frames, padding is needed to standardize input for the model.
- **Noisy Landmarks**: High variance in hand landmark coordinates (e.g., due to MediaPipe detection jitter) suggests normalization and smoothing to stabilize trajectories.
- **Missing Landmarks**: A high proportion of frames with zeroed hand landmarks (e.g., due to occlusion or detection failures) indicates preprocessing to handle gaps.

If none of these issues are present (e.g., all videos have ~45 frames, low variance, and few missing landmarks), preprocessing may be skipped, but applying it is generally safe and can improve model robustness.

### Notes
- **File Assumptions**: The code assumes landmarks are stored as `landmarks_complete.pkl` in subdirectories (e.g., `/content/drive/MyDrive/landmarks_data_corrected/pk-hfad-1_1/landmarks_complete.pkl`). If your files have a different name (e.g., `landmarks_raw.pkl`) or structure, update the `glob` pattern:
  ```python
  landmark_files = glob.glob(os.path.join(input_dir, '*/landmarks_raw.pkl'))
  ```
- **Customization**:
  - Adjust `target_frames=45` if you need a different frame count.
  - Modify `variance_threshold` (default 0.1) or `missing_threshold` (default 0.5) in `assess_preprocessing_need` based on your dataset’s characteristics.
  - Tune `sigma` in `smooth_landmarks` (default 1) to control smoothing intensity.
- **Performance**: The code is memory-efficient, loading and processing one video’s landmarks at a time. For very large datasets, you can add multiprocessing if needed:
  ```python
  from multiprocessing import Pool
  def process_file(landmark_file):
      preprocess_existing_landmarks(os.path.dirname(landmark_file))
  with Pool(4) as p:
      p.map(process_file, landmark_files)
  ```

````python
import numpy as np
import os
import pickle
import glob
from google.colab import drive
from scipy.ndimage import gaussian_filter1d



def assess_preprocessing_need(landmarks, target_frames=45, variance_threshold=0.1, missing_threshold=0.5):
    """
    Assess whether preprocessing is needed based on frame count, landmark variability, and missing landmarks.
    Returns a dictionary with analysis results and a boolean indicating if preprocessing is recommended.
    """
    analysis = {}

    # Check frame count
    frame_count = landmarks.shape[0]
    analysis['frame_count'] = frame_count
    analysis['frame_count_sufficient'] = frame_count >= target_frames * 0.8  # Allow 20% deviation

    # Check landmark variability (focus on hand landmarks)
    hand_landmarks = landmarks[:, 501:543, :2]  # x, y for right and left hands
    valid_landmarks = hand_landmarks[np.any(hand_landmarks != 0, axis=(1, 2))]
    if valid_landmarks.size > 0:
        variance = np.var(valid_landmarks, axis=0).mean()
        analysis['hand_variance'] = variance
        analysis['high_variability'] = variance > variance_threshold  # High variance suggests normalization/smoothing
    else:
        analysis['hand_variance'] = 0
        analysis['high_variability'] = False

    # Check for missing landmarks (zeros in hand landmarks)
    missing_ratio = np.mean(np.all(hand_landmarks == 0, axis=(1, 2)))
    analysis['missing_hand_ratio'] = missing_ratio
    analysis['excessive_missing'] = missing_ratio > missing_threshold  # Too many missing landmarks suggests preprocessing

    # Recommend preprocessing if any issue is detected
    preprocessing_needed = (
        not analysis['frame_count_sufficient'] or
        analysis['high_variability'] or
        analysis['excessive_missing']
    )

    return analysis, preprocessing_needed

def normalize_landmarks(landmarks):
    """
    Normalize landmarks per hand (right and left) and per axis (x, y only) to preserve gesture-specific dynamics.
    """
    normalized = landmarks.copy()
    # Normalize right hand (501:522) and left hand (522:543) separately
    for start, end in [(501, 522), (522, 543)]:
        hand = landmarks[:, start:end, :2]  # x, y only
        valid_mask = np.any(hand != 0, axis=(1, 2))  # Frames with non-zero landmarks
        if np.any(valid_mask):
            mean = hand[valid_mask].mean(axis=(0, 1), keepdims=True)
            std = hand[valid_mask].std(axis=(0, 1), keepdims=True) + 1e-8
            normalized[valid_mask, start:end, :2] = (hand[valid_mask] - mean) / std
    return normalized

def pad_landmarks(landmarks, target_frames=45):
    """
    Pad or truncate landmarks to target_frames, replicating the last valid frame for padding.
    """
    current_frames = landmarks.shape[0]
    if current_frames < target_frames:
        last_frame = landmarks[-1:]  # Repeat last frame
        pad = np.repeat(last_frame, target_frames - current_frames, axis=0)
        landmarks = np.vstack([landmarks, pad])
    return landmarks[:target_frames]

def smooth_landmarks(landmarks, sigma=1):
    """
    Apply Gaussian smoothing to reduce noise in landmark trajectories.
    """
    smoothed = landmarks.copy()
    valid_mask = np.any(landmarks != 0, axis=(1, 2))  # Frames with non-zero landmarks
    if np.any(valid_mask):
        smoothed[valid_mask] = gaussian_filter1d(landmarks[valid_mask], sigma=sigma, axis=0)
    return smoothed

def preprocess_existing_landmarks(input_dir, target_frames=45):
    """
    Load existing landmarks, assess preprocessing need, apply preprocessing if needed, and save results.
    """
    # Find all landmark files
    landmark_files = glob.glob(os.path.join(input_dir, '*/landmarks_complete.pkl'))

    for landmark_file in landmark_files:
        video_name = os.path.basename(os.path.dirname(landmark_file))
        preprocessed_output_file = os.path.join(input_dir, video_name, 'landmarks_preprocessed.pkl')

        # Skip if already preprocessed
        if os.path.exists(preprocessed_output_file):
            print(f"Skipping {video_name} (already preprocessed)")
            continue

        # Load landmarks
        try:
            with open(landmark_file, 'rb') as f:
                landmarks = pickle.load(f)
            landmarks = np.array(landmarks)
            print(f"Loaded landmarks for {video_name}: Shape {landmarks.shape}")
        except Exception as e:
            print(f"Error loading {video_name}: {e}")
            continue

        # Assess preprocessing need
        analysis, preprocessing_needed = assess_preprocessing_need(landmarks, target_frames)
        print(f"Preprocessing analysis for {video_name}: {analysis}")
        print(f"Preprocessing {'needed' if preprocessing_needed else 'not needed'}")

        # Apply preprocessing if needed
        if preprocessing_needed:
            landmarks = normalize_landmarks(landmarks)
            landmarks = smooth_landmarks(landmarks, sigma=1)
            landmarks = pad_landmarks(landmarks, target_frames)

        # Save preprocessed landmarks
        os.makedirs(os.path.dirname(preprocessed_output_file), exist_ok=True)
        with open(preprocessed_output_file, 'wb') as f:
            pickle.dump(landmarks, f)
        print(f"Saved preprocessed landmarks for {video_name}: Shape {landmarks.shape}, Hands: {np.any(landmarks[:, 501:543, :] != 0)}")

# Process existing landmarks
input_dir = '/content/Landmarks_WLASL'
preprocess_existing_landmarks(input_dir)
````

````python
import numpy as np
import pickle
import os
import json

def preprocess_landmarks_wlasl(folder_name, mapping_file, landmarks_base_dir="/content/Landmarks_WLASL", target_frames=45, min_frames=30, variance_threshold=0.05, missing_threshold=0.5):
    """
    Preprocess landmarks for a WLASL video, loading from landmarks_complete.pkl and saving to landmarks_preprocessed.pkl.
    Handles folder names with or without video ID (e.g., 'a lot_02123' or 'lion').

    Args:
        folder_name (str): Folder name (e.g., 'a lot_02123' or 'lion').
        mapping_file (str): Path to WLASL_v0.3.json for gloss and video ID mapping.
        landmarks_base_dir (str): Base directory for landmarks (default: '/content/Landmarks_WLASL').
        target_frames (int): Desired number of frames after preprocessing (default: 45).
        min_frames (int): Minimum acceptable frame count (default: 30).
        variance_threshold (float): Threshold for high hand variability (default: 0.05).
        missing_threshold (float): Threshold for excessive missing hand landmarks (default: 0.5).

    Returns:
        tuple: (preprocessed_landmarks (np.ndarray), analysis (dict), gloss (str), video_id (str or None))
    """
    # Load WLASL JSON
    with open(mapping_file, 'r') as f:
        wlasl_data = json.load(f)

    # Split folder name by last underscore to separate gloss and video ID
    parts = folder_name.rsplit('_', 1)
    if len(parts) == 2 and parts[1].isdigit():
        gloss_candidate = parts[0].replace('_', ' ')
        video_id = parts[1]
    else:
        gloss_candidate = folder_name.replace('_', ' ')
        video_id = None

    # Validate gloss and video ID against JSON
    gloss = None
    valid_video_id = None
    for entry in wlasl_data:
        if entry['gloss'].lower() == gloss_candidate.lower():
            gloss = entry['gloss']
            if video_id:
                for instance in entry['instances']:
                    if instance['video_id'] == video_id:
                        valid_video_id = video_id
                        break
            else:
                valid_video_id = None  # No video ID in folder name
            break

    if not gloss:
        raise ValueError(f"Invalid gloss '{gloss_candidate}' in folder '{folder_name}'")

    # Construct folder path
    folder_path = os.path.join(landmarks_base_dir, folder_name)

    # Load raw landmarks
    complete_landmarks_path = os.path.join(folder_path, 'landmarks_complete.pkl')
    if not os.path.exists(complete_landmarks_path):
        raise FileNotFoundError(f"Raw landmarks file not found: {complete_landmarks_path}")

    with open(complete_landmarks_path, 'rb') as f:
        landmarks_np = pickle.load(f)

    if not isinstance(landmarks_np, np.ndarray):
        landmarks_np = np.array(landmarks_np)

    if len(landmarks_np.shape) != 3 or landmarks_np.shape[1] != 543 or landmarks_np.shape[2] != 3:
        raise ValueError(f"Unexpected landmarks shape for {folder_name}: {landmarks_np.shape}")

    # Use folder_name for output to match provided format
    display_name = folder_name
    print(f"Loaded landmarks for {display_name}: Shape {landmarks_np.shape}")

    frame_count = landmarks_np.shape[0]
    hand_landmarks = landmarks_np[:, 468:543, :]
    hand_variance = np.var(hand_landmarks, axis=0).mean() if hand_landmarks.size > 0 else 0.0
    missing_hands = np.all(hand_landmarks == 0, axis=(1, 2))
    missing_hand_ratio = np.mean(missing_hands)

    analysis = {
        'frame_count': frame_count,
        'frame_count_sufficient': frame_count >= min_frames,
        'hand_variance': hand_variance,
        'high_variability': hand_variance > variance_threshold,
        'missing_hand_ratio': missing_hand_ratio,
        'excessive_missing': missing_hand_ratio > missing_threshold
    }
    print(f"Preprocessing analysis for {display_name}: {analysis}")

    preprocessing_needed = analysis['excessive_missing'] or analysis['high_variability'] or not analysis['frame_count_sufficient']
    print(f"Preprocessing {'needed' if preprocessing_needed else 'not needed'}")

    if preprocessing_needed:
        if frame_count > target_frames:
            landmarks_np = landmarks_np[:target_frames]
        elif frame_count < target_frames:
            padding = np.zeros((target_frames - frame_count, landmarks_np.shape[1], landmarks_np.shape[2]))
            landmarks_np = np.concatenate([landmarks_np, padding], axis=0)

        if analysis['excessive_missing']:
            for t in range(landmarks_np.shape[0]):
                if np.all(landmarks_np[t, 468:543, :] == 0):
                    if t > 0:
                        landmarks_np[t, 468:543, :] = landmarks_np[t-1, 468:543, :]

    has_hands = not np.all(hand_landmarks == 0)
    preprocessed_landmarks_path = os.path.join(folder_path, 'landmarks_preprocessed.pkl')
    with open(preprocessed_landmarks_path, 'wb') as f:
        pickle.dump(landmarks_np, f)
    print(f"Saved preprocessed landmarks for {display_name}: Shape {landmarks_np.shape}, Hands: {has_hands}")

    return landmarks_np, analysis, gloss, valid_video_id
````

````python
# Frame Count
import numpy as np
import pickle
import os
import json

def check_max_frame_count(mapping_file, landmarks_base_dir="/content/Landmarks_WLASL"):
    """
    Check the maximum frame count across all landmarks_complete.pkl files in the WLASL dataset.

    Args:
        mapping_file (str): Path to WLASL_v0.3.json for gloss and video ID mapping.
        landmarks_base_dir (str): Base directory for landmarks (default: '/content/Landmarks_WLASL').

    Returns:
        int: Maximum frame count found across all landmark files.
    """
    # Load WLASL JSON
    with open(mapping_file, 'r') as f:
        wlasl_data = json.load(f)

    # Create a set of valid glosses (case-insensitive)
    valid_glosses = {entry['gloss'].lower() for entry in wlasl_data}

    # Initialize maximum frame count
    max_frame_count = 0

    # Get all folder names in landmarks_base_dir
    folder_names = [f for f in os.listdir(landmarks_base_dir) if os.path.isdir(os.path.join(landmarks_base_dir, f))]

    for folder_name in folder_names:
        try:
            # Split folder name by last underscore to separate gloss and video ID
            parts = folder_name.rsplit('_', 1)
            if len(parts) == 2 and parts[1].isdigit():
                gloss_candidate = parts[0].replace('_', ' ')
                video_id = parts[1]
            else:
                gloss_candidate = folder_name.replace('_', ' ')
                video_id = None

            # Validate gloss against JSON
            if gloss_candidate.lower() not in valid_glosses:
                print(f"Skipping {folder_name}: Invalid gloss '{gloss_candidate}'")
                continue

            # Construct folder path
            folder_path = os.path.join(landmarks_base_dir, folder_name)

            # Load raw landmarks
            complete_landmarks_path = os.path.join(folder_path, 'landmarks_complete.pkl')
            if not os.path.exists(complete_landmarks_path):
                print(f"Skipping {folder_name}: Raw landmarks file not found")
                continue

            with open(complete_landmarks_path, 'rb') as f:
                landmarks = pickle.load(f)

            # Ensure landmarks are in numpy format
            if not isinstance(landmarks, np.ndarray):
                landmarks = np.array(landmarks)

            # Validate shape (expecting [frames, 543, 3])
            if len(landmarks.shape) != 3 or landmarks.shape[1] != 543 or landmarks.shape[2] != 3:
                print(f"Skipping {folder_name}: Unexpected landmarks shape {landmarks.shape}")
                continue

            # Get frame count
            frame_count = landmarks.shape[0]
            print(f"Processed {folder_name}: Frame count {frame_count}")

            # Update maximum frame count
            max_frame_count = max(max_frame_count, frame_count)

        except Exception as e:
            print(f"Error processing {folder_name}: {e}")
            continue

    print(f"\nMaximum frame count across all files: {max_frame_count}")
    return max_frame_count
````

````python
import numpy as np
import pickle
import os
import json
from glob import glob
from collections import Counter
import pandas as pd

def load_labels(mapping_file):
    """
    Load labels from WLASL_v0.3.json, mapping folder names to glosses.

    Args:
        mapping_file (str): Path to WLASL_v0.3.json.

    Returns:
        dict: Mapping of folder names (e.g., 'a lot_02123' or 'lion') to glosses.
    """
    with open(mapping_file, 'r') as f:
        wlasl_data = json.load(f)

    labels = {}
    for entry in wlasl_data:
        gloss = entry['gloss']
        for instance in entry['instances']:
            video_id = instance['video_id']
            # Create folder name with gloss and video ID (if applicable)
            folder_name = f"{gloss.replace(' ', '_')}_{video_id}" if video_id else gloss.replace(' ', '_')
            labels[folder_name] = gloss
        # Handle cases with gloss only (no video ID)
        folder_name = gloss.replace(' ', '_')
        if folder_name not in labels.values():
            labels[folder_name] = gloss

    return labels

def check_class_balance(labels):
    """
    Check class balance in labels.

    Args:
        labels (dict): Mapping of folder names to glosses.

    Returns:
        tuple: (distribution (dict), imbalance_detected (bool))
    """
    label_counts = Counter(labels.values())
    total = sum(label_counts.values())
    distribution = {label: count / total for label, count in label_counts.items()}
    imbalance_detected = max(label_counts.values()) / min(label_counts.values()) > 10  # Arbitrary threshold
    return distribution, imbalance_detected

def augment_landmarks(landmarks):
    """
    Apply data augmentation to landmarks (e.g., random scaling, noise).

    Args:
        landmarks (np.ndarray): Landmarks of shape (frames, keypoints, 3).

    Returns:
        np.ndarray: Augmented landmarks.
    """
    augmented = landmarks.copy()
    # Example augmentation: Add small random noise to coordinates
    noise = np.random.normal(0, 0.01, augmented.shape)
    augmented += noise
    # Ensure coordinates remain valid (e.g., clip to [0, 1] if normalized)
    augmented = np.clip(augmented, 0, 1)
    return augmented

def build_dataset(landmark_dir, label_file, target_frames=190, augment=True):
    """
    Build dataset from preprocessed landmark files and labels, with validation, augmentation, and balance check.
    Returns dataset, missing labels, and class weights.

    Args:
        landmark_dir (str): Directory containing landmark folders.
        label_file (str): Path to WLASL_v0.3.json.
        target_frames (int): Fixed frame count for landmarks (default: 190).
        augment (bool): Whether to apply data augmentation (default: True).

    Returns:
        tuple: (dataset (list of dicts), missing_labels (list), class_weights (dict))
    """
    # Load labels
    labels = load_labels(label_file)
    print(f"Loaded {len(labels)} labels: {list(labels.keys())[:5]}...")
    if not labels:
        print("Error: No labels loaded from label_file.")
        return [], [], {}

    # Check class balance
    distribution, imbalance_detected = check_class_balance(labels)
    print(f"Class balance: {distribution}, Imbalance detected: {imbalance_detected}")

    # Find preprocessed landmark files
    landmark_files = glob(os.path.join(landmark_dir, '*/landmarks_preprocessed.pkl'))
    video_names = [os.path.basename(os.path.dirname(f)) for f in landmark_files]
    print(f"Found {len(landmark_files)} landmark files: {video_names[:5]}...")

    # Validate labels
    missing_labels = [video_name for video_name in video_names if video_name not in labels]
    if missing_labels:
        print(f"Warning: No labels for {len(missing_labels)} videos: {missing_labels[:5]}...")
        pd.DataFrame({'video_name': missing_labels}).to_csv(
            os.path.join(landmark_dir, 'missing_labels.csv'), index=False
        )
        print(f"Saved missing labels to {os.path.join(landmark_dir, 'missing_labels.csv')}")

    # Build dataset
    dataset = []
    for landmark_file in landmark_files:
        video_name = os.path.basename(os.path.dirname(landmark_file))
        if video_name not in labels:
            continue

        try:
            with open(landmark_file, 'rb') as f:
                landmarks = pickle.load(f)
            landmarks = np.array(landmarks)
            print(f"Loaded landmarks for {video_name}: Shape {landmarks.shape}")

            # Validate shape
            if landmarks.shape != (target_frames, 543, 3):
                print(f"Skipping {video_name}: Expected shape ({target_frames}, 543, 3), got {landmarks.shape}")
                continue

            # Extract hand landmarks (indices 468-543, as in preprocessing)
            hand_landmarks = landmarks[:, 468:543, :]

            dataset.append({
                'video_name': video_name,
                'landmarks': hand_landmarks,
                'label': labels[video_name]
            })

            if augment:
                augmented_landmarks = augment_landmarks(hand_landmarks)
                dataset.append({
                    'video_name': f"{video_name}_aug",
                    'landmarks': augmented_landmarks,
                    'label': labels[video_name]
                })

            print(f"Added {video_name} (Shape: {hand_landmarks.shape}, Label: {labels[video_name]})")

        except Exception as e:
            print(f"Error loading {video_name}: {e}")
            continue

    print(f"Dataset size: {len(dataset)} videos (including {sum(1 for d in dataset if '_aug' in d['video_name'])} augmented)")

    # Compute class weights
    class_weights = {}
    if dataset:
        dataset_labels = [d['label'] for d in dataset]
        label_counts = Counter(dataset_labels)
        total = len(dataset_labels)
        class_weights = {label: total / (len(label_counts) * count) for label, count in label_counts.items()}
        print("Class weights:")
        for label, weight in sorted(class_weights.items()):
            print(f"  {label}: {weight:.3f}")

    return dataset, missing_labels, class_weights
````

````python
import torch
from torch.utils.data import Dataset, DataLoader
import numpy as np
import pickle
import os
import json

class WLASLDataset(Dataset):
    """
    PyTorch Dataset for WLASL landmarks and labels.
    """
    def __init__(self, dataset, label_to_idx):
        """
        Args:
            dataset (list): List of dicts from build_dataset, each with 'video_name', 'landmarks', 'label'.
            label_to_idx (dict): Mapping of glosses to integer indices.
        """
        self.dataset = dataset
        self.label_to_idx = label_to_idx

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        item = self.dataset[idx]
        landmarks = item['landmarks']  # Shape: (190, 75, 3)
        label = item['label']

        # Convert landmarks to tensor
        landmarks = torch.tensor(landmarks, dtype=torch.float32)  # Shape: (190, 75, 3)

        # Convert label to index
        label_idx = self.label_to_idx[label]

        return landmarks, label_idx

def create_data_loaders(dataset, batch_size=16, train_split=0.8, shuffle=True):
    """
    Create train and validation data loaders.

    Args:
        dataset (list): Dataset from build_dataset.
        batch_size (int): Batch size for data loaders.
        train_split (float): Fraction of data for training (rest for validation).
        shuffle (bool): Whether to shuffle data.

    Returns:
        tuple: (train_loader, val_loader, label_to_idx)
    """
    # Create label-to-index mapping
    labels = sorted(set(item['label'] for item in dataset))
    label_to_idx = {label: idx for idx, label in enumerate(labels)}

    # Split dataset into train and validation
    num_train = int(len(dataset) * train_split)
    train_dataset = dataset[:num_train]
    val_dataset = dataset[num_train:]

    # Create Dataset objects
    train_ds = WLASLDataset(train_dataset, label_to_idx)
    val_ds = WLASLDataset(val_dataset, label_to_idx)

    # Create DataLoaders
    train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=shuffle, num_workers=2)
    val_loader = DataLoader(val_ds, batch_size=batch_size, shuffle=False, num_workers=2)

    return train_loader, val_loader, label_to_idx
````

# Model Architecture

````python
import torch
import torch.nn as nn

class SignLanguageCNNLSTM(nn.Module):
    """
    CNN+LSTM model for sign language classification using hand landmarks.
    Input shape: (batch_size, frames=190, keypoints=75, channels=3)
    """
    def __init__(self, num_classes, hidden_size=128):
        super(SignLanguageCNNLSTM, self).__init__()

        # 3D CNN layers
        self.conv_layers = nn.Sequential(
            # Input: (batch_size, 3, 190, 75, 1)
            nn.Conv3d(in_channels=3, out_channels=16, kernel_size=(5, 3, 1), stride=1, padding=(2, 1, 0)),
            nn.ReLU(),
            nn.BatchNorm3d(16),
            nn.MaxPool3d(kernel_size=(2, 2, 1), stride=(2, 2, 1)),
            # Output: (batch_size, 16, 95, 37, 1)

            nn.Conv3d(16, 32, kernel_size=(5, 3, 1), stride=1, padding=(2, 1, 0)),
            nn.ReLU(),
            nn.BatchNorm3d(32),
            nn.MaxPool3d(kernel_size=(2, 2, 1), stride=(2, 2, 1)),
            # Output: (batch_size, 32, 47, 18, 1)

            nn.Conv3d(32, 64, kernel_size=(5, 3, 1), stride=1, padding=(2, 1, 0)),
            nn.ReLU(),
            nn.BatchNorm3d(64),
            nn.MaxPool3d(kernel_size=(2, 2, 1), stride=(2, 2, 1)),
            # Output: (batch_size, 64, 23, 9, 1)
        )

        # Calculate flattened size after CNN
        self.flatten_size = 64 * 23 * 9 * 1  # Adjust based on CNN output

        # LSTM layer
        self.lstm = nn.LSTM(input_size=self.flatten_size // 23,  # Features per frame
                           hidden_size=hidden_size,
                           num_layers=1,
                           batch_first=True)

        # Fully connected layers
        self.fc_layers = nn.Sequential(
            nn.Linear(hidden_size, 128),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        """
        Forward pass.
        Input shape: (batch_size, 190, 75, 3)
        Output shape: (batch_size, num_classes)
        """
        # Reshape input: (batch_size, 190, 75, 3) -> (batch_size, 3, 190, 75, 1)
        batch_size = x.size(0)
        x = x.permute(0, 3, 1, 2).unsqueeze(-1)  # (batch_size, 3, 190, 75, 1)

        # CNN
        x = self.conv_layers(x)  # (batch_size, 64, 23, 9, 1)

        # Reshape for LSTM: (batch_size, 23, 64*9*1)
        x = x.squeeze(-1)  # (batch_size, 64, 23, 9)
        x = x.view(batch_size, 23, -1)  # (batch_size, 23, 64*9)

        # LSTM
        x, (hn, cn) = self.lstm(x)  # (batch_size, 23, hidden_size)
        x = x[:, -1, :]  # Take last time step: (batch_size, hidden_size)

        # FC layers
        x = self.fc_layers(x)  # (batch_size, num_classes)
        return x
````

# Model Training

````python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim.lr_scheduler import ReduceLROnPlateau
from sklearn.metrics import accuracy_score, f1_score
import numpy as np
import os
import pickle
from tqdm import tqdm

# Assume previous functions are available
# from your_previous_module import build_dataset, create_data_loaders, SignLanguageCNNLSTM  # Replace with actual import

def train_model(train_loader, val_loader, class_weights, num_classes=2000, num_epochs=50, batch_size=16, learning_rate=1e-3, device='cuda', save_path='best_model.pth'):
    """
    Train the SignLanguageCNNLSTM model and save the best checkpoint.

    Args:
        train_loader (DataLoader): Training data loader.
        val_loader (DataLoader): Validation data loader.
        class_weights (dict): Class weights from build_dataset.
        num_classes (int): Number of gloss classes (default: 2000).
        num_epochs (int): Number of training epochs (default: 20).
        batch_size (int): Batch size (default: 16).
        learning_rate (float): Initial learning rate (default: 1e-3).
        device (str): Device ('cuda' or 'cpu').
        save_path (str): Path to save best model checkpoint.

    Returns:
        dict: Training history (train/val loss, accuracy, F1-score).
    """
    # Initialize model
    model = SignLanguageCNNLSTM(num_classes=num_classes).to(device)
    print(model)

    # Convert class weights to tensor
    class_weights_tensor = torch.tensor([class_weights.get(i, 1.0) for i in range(num_classes)], dtype=torch.float32).to(device)

    # Loss and optimizer
    criterion = nn.CrossEntropyLoss(weight=class_weights_tensor)
    optimizer = optim.AdamW(model.parameters(), lr=learning_rate)
    scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=3, verbose=True)

    # Training history
    history = {
        'train_loss': [], 'val_loss': [],
        'train_acc': [], 'val_acc': [],
        'train_f1': [], 'val_f1': []
    }

    best_val_loss = float('inf')
    patience = 5
    early_stop_counter = 0

    for epoch in range(num_epochs):
        # Training phase
        model.train()
        train_loss, train_preds, train_labels = 0.0, [], []

        for landmarks, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs} [Train]"):
            landmarks, labels = landmarks.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(landmarks)  # Shape: (batch_size, num_classes)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            train_loss += loss.item() * landmarks.size(0)
            _, preds = torch.max(outputs, 1)
            train_preds.extend(preds.cpu().numpy())
            train_labels.extend(labels.cpu().numpy())

        train_loss /= len(train_loader.dataset)
        train_acc = accuracy_score(train_labels, train_preds)
        train_f1 = f1_score(train_labels, train_preds, average='weighted')

        # Validation phase
        model.eval()
        val_loss, val_preds, val_labels = 0.0, [], []

        with torch.no_grad():
            for landmarks, labels in tqdm(val_loader, desc=f"Epoch {epoch+1}/{num_epochs} [Val]"):
                landmarks, labels = landmarks.to(device), labels.to(device)

                outputs = model(landmarks)
                loss = criterion(outputs, labels)

                val_loss += loss.item() * landmarks.size(0)
                _, preds = torch.max(outputs, 1)
                val_preds.extend(preds.cpu().numpy())
                val_labels.extend(labels.cpu().numpy())

        val_loss /= len(val_loader.dataset)
        val_acc = accuracy_score(val_labels, val_preds)
        val_f1 = f1_score(val_labels, val_preds, average='weighted')

        # Update history
        history['train_loss'].append(train_loss)
        history['val_loss'].append(val_loss)
        history['train_acc'].append(train_acc)
        history['val_acc'].append(val_acc)
        history['train_f1'].append(train_f1)
        history['val_f1'].append(val_f1)

        # Print epoch results
        print(f"Epoch {epoch+1}/{num_epochs}:")
        print(f"  Train Loss: {train_loss:.4f}, Acc: {train_acc:.4f}, F1: {train_f1:.4f}")
        print(f"  Val Loss: {val_loss:.4f}, Acc: {val_acc:.4f}, F1: {val_f1:.4f}")

        # Learning rate scheduling
        scheduler.step(val_loss)

        # Save best model
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            torch.save({
                'epoch': epoch + 1,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'val_loss': val_loss,
                'history': history
            }, save_path)
            print(f"Saved best model to {save_path} (Val Loss: {val_loss:.4f})")
            early_stop_counter = 0
        else:
            early_stop_counter += 1

        # Early stopping
        if early_stop_counter >= patience:
            print(f"Early stopping triggered after {epoch+1} epochs")
            break

    return history

# Example usage
def main():
    # Parameters
    landmark_dir = '/content/drive/MyDrive/Datasets_ISL-CSLRT-PSL/Processed_Landmarks_WLASL'
    label_file = '/content/drive/MyDrive/Datasets_ISL-CSLRT-PSL/WLASL_v0.3.json'
    batch_size = 16
    num_epochs = 20
    learning_rate = 1e-3
    save_path = '/content/best_sign_language_model_WLASL.pth'
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    # Build dataset
    dataset, missing_labels, class_weights = build_dataset(
        landmark_dir, label_file, target_frames=190, augment=True
    )

    # Create data loaders
    train_loader, val_loader, label_to_idx = create_data_loaders(
        dataset, batch_size=batch_size, train_split=0.8, shuffle=True
    )

    # Train model
    history = train_model(
        train_loader, val_loader, class_weights,
        num_classes=len(label_to_idx), num_epochs=num_epochs,
        batch_size=batch_size, learning_rate=learning_rate,
        device=device, save_path=save_path
    )

    # Save training history
    with open(os.path.join(os.path.dirname(save_path), 'training_history.pkl'), 'wb') as f:
        pickle.dump(history, f)
    print("Training completed. History saved.")

if __name__ == '__main__':
    main()
````

# Map LandMarks folder Final

````python
import os
import json
from pathlib import Path
from glob import glob
from collections import Counter, defaultdict
import pandas as pd

def get_landmark_folders(landmark_dir):
    landmark_dir = Path(landmark_dir)
    folders = [
        f.name for f in landmark_dir.iterdir()
        if f.is_dir() and (f / 'landmarks_preprocessed.pkl').exists()
    ]
    print(f"Found {len(folders)} landmark folders:")
    for folder in sorted(folders)[:10]:
        print(f"  {folder}")
    if len(folders) > 10:
        print("  ...")
    return folders

def search_landmark_folders(base_dir):
    base_dir = Path(base_dir)
    folders = [
        str(f.parent) for f in base_dir.rglob('landmarks_preprocessed.pkl')
    ]
    print(f"Found {len(folders)} landmark folders across all directories:")
    for f in folders[:10]:
        print(f"  {f}")
    if len(folders) > 10:
        print("  ...")
    return folders

def extract_gloss_from_folder(folder_name):
    """Extract gloss from folder name, ignoring video ID."""
    parts = folder_name.rsplit('_', 1)
    gloss = parts[0].replace('_', ' ').lower()
    return gloss

def extract_wlasl_names(json_path):
    with open(json_path, 'r') as f:
        wlasl_data = json.load(f)

    expected_folders = set()
    gloss_to_videos = {}
    gloss_counts = Counter()

    for entry in wlasl_data:
        gloss = entry['gloss']
        gloss_normalized = gloss.replace(' ', '_').lower()
        video_ids = [inst['video_id'] for inst in entry['instances']]
        gloss_counts[gloss] = len(video_ids)
        gloss_to_videos[gloss_normalized] = video_ids
        expected_folders.add(gloss_normalized)
        for instance in entry['instances']:
            video_id = instance['video_id']
            folder_name = f"{gloss_normalized}_{video_id}"
            expected_folders.add(folder_name)

    print(f"Total glosses: {len(gloss_counts)}")
    print("Gloss distribution (videos per gloss):")
    for gloss, count in sorted(gloss_counts.items())[:10]:
        print(f"  {gloss}: {count}")
    if len(gloss_counts) > 10:
        print("  ...")
    print(f"Sample expected folder names: {list(expected_folders)[:10]}")

    return expected_folders, gloss_to_videos, gloss_counts

def load_labels(mapping_file, bilingual_mapping_file, landmark_folders):
    """
    Load labels for all glosses from landmark folders, prioritizing bilingual glosses.
    """
    # Load WLASL JSON
    with open(mapping_file, 'r') as f:
        wlasl_data = json.load(f)

    # Load bilingual mapping
    with open(bilingual_mapping_file, 'r') as f:
        bilingual_data = json.load(f)

    # Extract gloss counts and video IDs from WLASL
    gloss_counts = Counter()
    gloss_to_videos = {}
    for entry in wlasl_data:
        gloss = entry['gloss']
        gloss_normalized = gloss.replace(' ', '_').lower()
        video_ids = [inst['video_id'] for inst in entry['instances']]
        gloss_counts[gloss] = len(video_ids)
        gloss_to_videos[gloss_normalized] = video_ids

    # Get bilingual glosses
    bilingual_glosses = {entry['word'].lower() for entry in bilingual_data['sign_language_mapping']}
    print(f"Found {len(bilingual_glosses)} bilingual glosses: {list(bilingual_glosses)[:10]}...")

    # Extract glosses from all landmark folders
    folder_glosses = set(extract_gloss_from_folder(folder) for folder in landmark_folders)
    print(f"Found {len(folder_glosses)} unique glosses from landmark folders: {list(folder_glosses)[:10]}...")

    # Map folders to their glosses
    folder_to_gloss = {folder: extract_gloss_from_folder(folder) for folder in landmark_folders}

    # Identify available glosses that have landmark folders and are in WLASL
    available_glosses = set()
    for gloss in folder_glosses:
        for entry in wlasl_data:
            if gloss == entry['gloss'].lower():
                available_glosses.add(entry['gloss'])
                break

    # Check for common glosses between bilingual and available glosses
    common_bilingual_glosses = bilingual_glosses & {g.lower() for g in available_glosses}
    print(f"Found {len(common_bilingual_glosses)} common bilingual glosses with landmark folders: {list(common_bilingual_glosses)[:10]}...")

    # Select all available glosses (no max_glosses cap)
    selected_glosses = sorted(
        {g.lower() for g in available_glosses},
        key=lambda gloss: (gloss in common_bilingual_glosses, gloss_counts.get(gloss.capitalize(), 0)),
        reverse=True
    )

    # Build labels for all selected glosses
    labels = {}
    folder_to_label = {}
    for entry in wlasl_data:
        gloss = entry['gloss']
        if gloss.lower() in selected_glosses:
            gloss_normalized = gloss.replace(' ', '_').lower()
            video_ids = [inst['video_id'] for inst in entry['instances']]
            for video_id in video_ids:
                folder_name = f"{gloss_normalized}_{video_id}"
                labels[folder_name] = gloss
                folder_to_label[folder_name] = gloss
            labels[gloss_normalized] = gloss

    # Assign labels to all landmark folders
    for folder in landmark_folders:
        gloss = folder_to_gloss[folder]
        if gloss in selected_glosses and folder not in folder_to_label:
            folder_to_label[folder] = next(g for g in labels.values() if g.lower() == gloss)

    print(f"Selected {len(selected_glosses)} glosses with landmark folders")
    print(f"Bilingual glosses included: {len(bilingual_glosses & set(selected_glosses))}")
    print("Gloss distribution in JSON:")
    for gloss in sorted(selected_glosses)[:10]:
        print(f"  {gloss}: {gloss_counts.get(gloss.capitalize(), 0)}")
    if len(selected_glosses) > 10:
        print("  ...")
    print(f"Sample label mappings: {list(folder_to_label.items())[:10]}")
    return folder_to_label, gloss_to_videos

def check_missing_folders(landmark_folders, gloss_to_videos):
    missing_videos = {}
    folder_glosses = {folder: extract_gloss_from_folder(folder) for folder in landmark_folders}
    for gloss, video_ids in gloss_to_videos.items():
        expected_folders = [f"{gloss}_{vid}" for vid in video_ids]
        found_folders = [f for f in landmark_folders if folder_glosses[f] == gloss.replace('_', ' ').lower()]
        missing = [vid for vid in video_ids if f"{gloss}_{vid}" not in landmark_folders]
        if missing:
            missing_videos[gloss] = missing

    print(f"Glosses with missing videos: {len(missing_videos)}")
    for gloss, vids in sorted(missing_videos.items())[:5]:
        print(f"  {gloss}: {vids}")
    if len(missing_videos) > 5:
        print("  ...")

    return missing_videos

def compare_folders(landmark_folders, expected_folders, gloss_to_videos, labels):
    matches = []
    mismatches = []
    unmatched = []
    folder_to_gloss = {folder: extract_gloss_from_folder(folder) for folder in landmark_folders}

    for folder in landmark_folders:
        gloss = folder_to_gloss[folder]
        if folder in labels:
            matches.append(folder)
        elif gloss in {g.lower() for g in labels.values()}:
            gloss_normalized = gloss.replace(' ', '_').lower()
            video_ids = gloss_to_videos.get(gloss_normalized, [])
            if video_ids:
                mismatches.append((folder, f"{gloss_normalized}_{video_ids[0]}"))
            else:
                unmatched.append(folder)
        else:
            unmatched.append(folder)

    gloss_to_folders = defaultdict(list)
    for folder, gloss in folder_to_gloss.items():
        if gloss in {g.lower() for g in labels.values()}:
            gloss_to_folders[next(g for g in labels.values() if g.lower() == gloss)].append(folder)

    print(f"Gloss grouping (sample):")
    for gloss, folders in sorted(gloss_to_folders.items())[:5]:
        print(f"  {gloss}: {folders}")
    if len(gloss_to_folders) > 5:
        print("  ...")

    print(f"Matches: {len(matches)} folders")
    for folder in matches[:5]:
        print(f"  {folder}")
    if len(matches) > 5:
        print("  ...")

    print(f"Mismatches (gloss-only folders to rename): {len(mismatches)} folders")
    for old, new in mismatches[:5]:
        print(f"  {old} → {new}")
    if len(mismatches) > 5:
        print("  ...")

    print(f"Unmatched: {len(unmatched)} folders")
    for folder in unmatched[:5]:
        print(f"  {folder}")
    if len(unmatched) > 5:
        print("  ...")

    return matches, mismatches, unmatched

def rename_folders(landmark_dir, mismatches, unmatched):
    landmark_dir = Path(landmark_dir)
    renamed = []

    for old_name, new_name in mismatches:
        old_path = landmark_dir / old_name
        new_path = landmark_dir / new_name
        if old_path.exists() and not new_path.exists():
            old_path.rename(new_path)
            renamed.append((old_name, new_name))
        else:
            if not old_path.exists():
                print(f"Warning: {old_name} does not exist")
            if new_path.exists():
                print(f"Warning: {new_name} already exists, skipping rename")

    if unmatched:
        pd.DataFrame({'unmatched_folder': unmatched}).to_csv(
            landmark_dir / 'unmatched_folders_final.csv', index=False
        )
        print(f"Saved {len(unmatched)} unmatched folders to {landmark_dir / 'unmatched_folders_final.csv'}")

    print(f"Renamed {len(renamed)} folders:")
    for old, new in renamed[:5]:
        print(f"  {old} → {new}")
    if len(renamed) > 5:
        print("  ...")

    return renamed

def test_build_dataset(landmark_dir, label_file, bilingual_mapping_file, landmark_folders, min_samples_per_class=1):
    """
    Test folder-to-gloss mapping without processing landmarks.
    """
    labels, gloss_to_videos = load_labels(label_file, bilingual_mapping_file, landmark_folders)

    landmark_files = glob(os.path.join(landmark_dir, '*/landmarks_preprocessed.pkl'))
    video_names = [os.path.basename(os.path.dirname(f)) for f in landmark_files]
    print(f"\nFound {len(landmark_files)} landmark files: {video_names[:5]}...")

    missing_labels = [video_name for video_name in video_names if video_name not in labels]
    if missing_labels:
        print(f"\nWarning: No labels found for {len(missing_labels)} videos: {missing_labels[:5]}...")
        pd.DataFrame({'video_name': missing_labels}).to_csv(
            os.path.join(landmark_dir, 'missing_labels.csv'), index=False
        )

    dataset = []
    for video_name in video_names:
        if video_name in labels:
            dataset.append({
                'video_name': video_name,
                'label': labels[video_name]
            })

    label_counts = Counter(d['label'] for d in dataset)
    valid_glosses = {gloss for gloss, count in label_counts.items() if count >= min_samples_per_class}
    dataset = [d for d in dataset if d['label'] in valid_glosses]

    print(f"\nFiltered to {len(valid_glosses)} glosses with ≥{min_samples_per_class} samples in dataset")
    print(f"Dataset size: {len(dataset)} videos")
    print("Class distribution:")
    for label, count in sorted(label_counts.items())[:10]:
        print(f"  {label}: {count}")
    if len(label_counts) > 10:
        print("  ...")
    print(f"Number of classes: {len(label_counts)}")

    return dataset, missing_labels

if __name__ == "__main__":
    landmark_dir = '/content/drive/MyDrive/Datasets_ISL-CSLRT-PSL/Processed_Landmarks_WLASL'
    label_file = '/content/drive/MyDrive/Datasets_ISL-CSLRT-PSL/WLASL_v0.3.json'
    bilingual_mapping_file = '/content/drive/MyDrive/Datasets_ISL-CSLRT-PSL/bilingual_sign_mapping.json'
    base_dir = '/content/drive/MyDrive/Datasets_ISL-CSLRT-PSL'

    print("Searching for all landmark folders...")
    search_landmark_folders(base_dir)

    landmark_folders = get_landmark_folders(landmark_dir)
    expected_folders, gloss_to_videos, _ = extract_wlasl_names(label_file)
    labels, gloss_to_videos = load_labels(label_file, bilingual_mapping_file, landmark_folders)
    missing_videos = check_missing_folders(landmark_folders, gloss_to_videos)
    matches, mismatches, unmatched = compare_folders(landmark_folders, expected_folders, gloss_to_videos, labels)
    renamed = rename_folders(landmark_dir, mismatches, unmatched)
    dataset, missing_labels = test_build_dataset(landmark_dir, label_file, bilingual_mapping_file, landmark_folders, min_samples_per_class=1)
````

````python
# Augmentation and saving
import os
import pickle
import numpy as np
from pathlib import Path

def augment_landmarks(landmarks, rotation_angle=10, scale_factor=1.1, noise_std=0.01, frame_drop_prob=0.1):
    """
    Augment landmark data with spatial and temporal transformations.
    Args:
        landmarks: NumPy array of shape [frames, keypoints, 3] (x, y, z coordinates).
        rotation_angle: Max rotation angle in degrees.
        scale_factor: Max scaling factor.
        noise_std: Standard deviation for Gaussian noise.
        frame_drop_prob: Probability of dropping a frame for temporal augmentation.
    Returns:
        Augmented landmarks array of same shape.
    """
    augmented = landmarks.copy()
    frames, keypoints, coords = augmented.shape

    # Spatial augmentation: Rotation and scaling (applied to x, y coordinates)
    theta = np.radians(np.random.uniform(-rotation_angle, rotation_angle))
    rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])
    scale = np.random.uniform(0.9, scale_factor)

    # Apply rotation and scaling to x, y (leave z unchanged)
    xy = augmented[:, :, :2]  # [frames, keypoints, 2]
    xy = np.dot(xy, rotation_matrix) * scale
    augmented[:, :, :2] = xy

    # Add Gaussian noise to all coordinates
    augmented += np.random.normal(0, noise_std, augmented.shape)

    # Temporal augmentation: Random frame dropping
    keep_mask = np.random.uniform(0, 1, frames) > frame_drop_prob
    if keep_mask.sum() > 0:  # Ensure at least one frame remains
        augmented = augmented[keep_mask]
        # Pad back to original frame count if needed
        if len(augmented) < frames:
            pad_size = frames - len(augmented)
            augmented = np.pad(augmented, ((0, pad_size), (0, 0), (0, 0)), mode='edge')
    else:
        # If all frames dropped, keep one random frame
        random_frame = augmented[np.random.randint(frames)][None, :, :]
        augmented = np.repeat(random_frame, frames, axis=0)

    return augmented

def augment_dataset(landmark_dir, dataset, num_augmentations=4):
    """
    Augment the dataset by applying transformations to landmark files, skipping existing augmented files.
    Args:
        landmark_dir: Path to landmark directory (e.g., '/content/drive/.../Processed_Landmarks_WLASL').
        dataset: List of dicts [{'video_name': ..., 'label': ...}, ...].
        num_augmentations: Number of augmented versions per video (e.g., 4).
    Returns:
        Updated dataset with original and augmented entries.
    """
    landmark_dir = Path(landmark_dir)
    augmented_dataset = dataset.copy()
    skipped_videos = []

    for entry in dataset:
        video_name = entry['video_name']
        label = entry['label']
        landmark_path = landmark_dir / video_name / 'landmarks_preprocessed.pkl'

        if not landmark_path.exists():
            print(f"Warning: {landmark_path} does not exist, skipping augmentation")
            continue

        # Check if all augmented versions already exist
        all_augmented_exist = True
        for i in range(num_augmentations):
            aug_video_name = f"{video_name}_aug{i}"
            aug_dir = landmark_dir / aug_video_name
            aug_path = aug_dir / 'landmarks_preprocessed.pkl'
            if not aug_path.exists():
                all_augmented_exist = False
                break

        if all_augmented_exist:
            print(f"Skipping augmentation for {video_name}: all {num_augmentations} augmented versions exist")
            skipped_videos.append(video_name)
            # Ensure augmented entries are in the dataset
            for i in range(num_augmentations):
                aug_video_name = f"{video_name}_aug{i}"
                if not any(d['video_name'] == aug_video_name for d in augmented_dataset):
                    augmented_dataset.append({
                        'video_name': aug_video_name,
                        'label': label
                    })
            continue

        # Load landmarks
        with open(landmark_path, 'rb') as f:
            landmarks = pickle.load(f)

        # Convert to NumPy array if not already
        if isinstance(landmarks, dict):
            # Combine face, hands, pose into a single array
            landmarks = np.concatenate([
                np.array(landmarks.get('face', np.zeros((30, 468, 3)))),
                np.array(landmarks.get('left_hand', np.zeros((30, 21, 3)))),
                np.array(landmarks.get('right_hand', np.zeros((30, 21, 3)))),
                np.array(landmarks.get('pose', np.zeros((30, 33, 3))))
            ], axis=1)  # Shape: [frames, 543, 3]
        elif not isinstance(landmarks, np.ndarray):
            print(f"Warning: Unexpected landmark format in {video_name}, skipping")
            continue

        # Apply augmentations
        for i in range(num_augmentations):
            aug_video_name = f"{video_name}_aug{i}"
            aug_dir = landmark_dir / aug_video_name
            aug_path = aug_dir / 'landmarks_preprocessed.pkl'

            # Skip if augmented file already exists
            if aug_path.exists():
                print(f"Skipping {aug_video_name}: file already exists")
                # Ensure entry is in dataset
                if not any(d['video_name'] == aug_video_name for d in augmented_dataset):
                    augmented_dataset.append({
                        'video_name': aug_video_name,
                        'label': label
                    })
                continue

            # Create output directory
            os.makedirs(aug_dir, exist_ok=True)

            # Augment landmarks
            aug_landmarks = augment_landmarks(landmarks)

            # Save augmented landmarks
            with open(aug_path, 'wb') as f:
                pickle.dump(aug_landmarks, f)

            # Add to dataset
            augmented_dataset.append({
                'video_name': aug_video_name,
                'label': label
            })

    print(f"Augmented dataset size: {len(augmented_dataset)} videos")
    print(f"Skipped augmentation for {len(skipped_videos)} videos")
    print("Sample augmented entries:")
    for entry in augmented_dataset[-10:]:
        print(f"  {entry}")

    return augmented_dataset

if __name__ == "__main__":
    # Example usage (replace with your actual dataset)
    landmark_dir = '/content/drive/MyDrive/Datasets_ISL-CSLRT-PSL/Processed_Landmarks_WLASL'
    # Assume dataset from map_landmark_folders_bilingual.py
    dataset = [
        {'video_name': 'coconut_65374', 'label': 'coconut'}, {'video_name': 'fruit_23708', 'label': 'fruit'}, {'video_name': 'e_18107', 'label': 'e'}, {'video_name': 'pants_66261', 'label': 'pants'}, {'video_name': 'wait_62077', 'label': 'wait'}, {'video_name': 'embarrass_18887', 'label': 'embarrass'}, {'video_name': 'promise_44896', 'label': 'promise'}, {'video_name': 'river_48249', 'label': 'river'}, {'video_name': 'crocodile_65425', 'label': 'crocodile'}, {'video_name': 'coffee_11552', 'label': 'coffee'}, {'video_name': 'frog_23626', 'label': 'frog'}, {'video_name': 'your_64423', 'label': 'your'}, {'video_name': 'm_34336', 'label': 'm'}, {'video_name': 'judge_31251', 'label': 'judge'}, {'video_name': 'flower_22537', 'label': 'flower'}, {'video_name': 'excuse_20197', 'label': 'excuse'}, {'video_name': 'woman_63662', 'label': 'woman'}, {'video_name': 'pass_66267', 'label': 'pass'}, {'video_name': 'wedding_62684', 'label': 'wedding'}, {'video_name': 'shop_66468', 'label': 'shop'}, {'video_name': 'nice_38281', 'label': 'nice'}, {'video_name': 'b_66040', 'label': 'b'}, {'video_name': 's_48933', 'label': 's'}, {'video_name': 'america_02348', 'label': 'america'}, {'video_name': 'bed_05628', 'label': 'bed'}, {'video_name': 'tomorrow_58804', 'label': 'tomorrow'}, {'video_name': 'try_59847', 'label': 'try'}, {'video_name': 'night_38329', 'label': 'night'}, {'video_name': 'month_36732', 'label': 'month'}, {'video_name': 'rain_45965', 'label': 'rain'}, {'video_name': 'university_60759', 'label': 'university'}, {'video_name': 'bored_65229', 'label': 'bored'}, {'video_name': 'beg_05756', 'label': 'beg'}, {'video_name': 'i_28557', 'label': 'i'}, {'video_name': 'prepare_44267', 'label': 'prepare'}, {'video_name': 'chicken_10395', 'label': 'chicken'}, {'video_name': 'w_62032', 'label': 'w'}, {'video_name': 'christian_10685', 'label': 'christian'}, {'video_name': 'break_07608', 'label': 'break'}, {'video_name': 'birthday_06355', 'label': 'birthday'}, {'video_name': 'horse_27983', 'label': 'horse'}, {'video_name': 't_56546', 'label': 't'}, {'video_name': 'company_12099', 'label': 'company'}, {'video_name': 'science_49627', 'label': 'science'}, {'video_name': 'lazy_32509', 'label': 'lazy'}, {'video_name': 'mountain_37013', 'label': 'mountain'}, {'video_name': 'alarm_01796', 'label': 'alarm'}, {'video_name': 'busy_08410', 'label': 'busy'}, {'video_name': 'middle_35962', 'label': 'middle'}, {'video_name': 'p_40701', 'label': 'p'}, {'video_name': 'art_03469', 'label': 'art'}, {'video_name': 'week_62728', 'label': 'week'}, {'video_name': 'for_22836', 'label': 'for'}, {'video_name': 'lie_66067', 'label': 'lie'}, {'video_name': 'n_37512', 'label': 'n'}, {'video_name': 'fool_65754', 'label': 'fool'}, {'video_name': 'meeting_35524', 'label': 'meeting'}, {'video_name': 'stubborn_55312', 'label': 'stubborn'}, {'video_name': 'weak_62584', 'label': 'weak'}, {'video_name': 'always_02227', 'label': 'always'}, {'video_name': 'easy_18301', 'label': 'easy'}, {'video_name': 'age_01482', 'label': 'age'}, {'video_name': 'ignore_28772', 'label': 'ignore'}, {'video_name': 'law_66017', 'label': 'law'}, {'video_name': 'walk_62152', 'label': 'walk'}, {'video_name': 'v_61130', 'label': 'v'}, {'video_name': 'lemon_32806', 'label': 'lemon'}, {'video_name': 'talk_56800', 'label': 'talk'}, {'video_name': 'tree_59540', 'label': 'tree'}, {'video_name': 'there_57815', 'label': 'there'}, {'video_name': 'sure_56025', 'label': 'sure'}, {'video_name': 'grapes_25481', 'label': 'grapes'}, {'video_name': 'potato_43922', 'label': 'potato'}, {'video_name': 'aunt_04165', 'label': 'aunt'}, {'video_name': 'english_19230', 'label': 'english'}, {'video_name': 'rat_46155', 'label': 'rat'}, {'video_name': 'story_54991', 'label': 'story'}, {'video_name': 'gold_24978', 'label': 'gold'}, {'video_name': 'second_66439', 'label': 'second'}, {'video_name': 'bread_07593', 'label': 'bread'}, {'video_name': 'heaven_27085', 'label': 'heaven'}, {'video_name': 'k_31436', 'label': 'k'}, {'video_name': 'water_62479', 'label': 'water'}, {'video_name': 'society_52990', 'label': 'society'}, {'video_name': 'word_63750', 'label': 'word'}, {'video_name': 'monkey_36677', 'label': 'monkey'}, {'video_name': 'closet_11292', 'label': 'closet'}, {'video_name': 'giraffe_24580', 'label': 'giraffe'}, {'video_name': 'iron_65964', 'label': 'iron'}, {'video_name': 'fail_20792', 'label': 'fail'}, {'video_name': 'grandfather_25404', 'label': 'grandfather'}, {'video_name': 'daughter_14748', 'label': 'daughter'}, {'video_name': 'quick_45739', 'label': 'quick'}, {'video_name': 'grandmother_25428', 'label': 'grandmother'}, {'video_name': 'd_66042', 'label': 'd'}, {'video_name': 'welcome_62826', 'label': 'welcome'}, {'video_name': 'hospital_68404', 'label': 'hospital'}, {'video_name': 'newspaper_38176', 'label': 'newspaper'}, {'video_name': 'minute_66128', 'label': 'minute'}, {'video_name': 'ill_28823', 'label': 'ill'}, {'video_name': 'tomato_58782', 'label': 'tomato'}, {'video_name': 'fine_65717', 'label': 'fine'}, {'video_name': 'loan_33607', 'label': 'loan'}, {'video_name': 'date_65442', 'label': 'date'}, {'video_name': 'street_55145', 'label': 'street'}, {'video_name': 'famous_21008', 'label': 'famous'}, {'video_name': 'q_45595', 'label': 'q'}, {'video_name': 'august_04154', 'label': 'august'}, {'video_name': 'teacher_57037', 'label': 'teacher'}, {'video_name': 'happy_26506', 'label': 'happy'}, {'video_name': 'serve_50528', 'label': 'serve'}, {'video_name': 'use_61031', 'label': 'use'}, {'video_name': 'butter_08439', 'label': 'butter'}, {'video_name': 'goat_24883', 'label': 'goat'}, {'video_name': 'punish_45362', 'label': 'punish'}, {'video_name': 'j_30820', 'label': 'j'}, {'video_name': 'angel_02539', 'label': 'angel'}, {'video_name': 'admire_65020', 'label': 'admire'}, {'video_name': 'become_05614', 'label': 'become'}, {'video_name': 'owl_40658', 'label': 'owl'}, {'video_name': 'free_23395', 'label': 'free'}, {'video_name': 'stairs_66552', 'label': 'stairs'}, {'video_name': 'elephant_18769', 'label': 'elephant'}, {'video_name': 'television_57260', 'label': 'television'}, {'video_name': 'u_60338', 'label': 'u'}, {'video_name': 'result_47729', 'label': 'result'}, {'video_name': 'upset_60947', 'label': 'upset'}, {'video_name': 'information_29631', 'label': 'information'}, {'video_name': 'interesting_68308', 'label': 'interesting'}, {'video_name': 'exercise_65657', 'label': 'exercise'}, {'video_name': 'f_20670', 'label': 'f'}, {'video_name': 'o_39147', 'label': 'o'}, {'video_name': 'friday_23541', 'label': 'friday'}, {'video_name': 'cold_65377', 'label': 'cold'}, {'video_name': 'and_02517', 'label': 'and'}, {'video_name': 'g_23959', 'label': 'g'}, {'video_name': 'motorcycle_36991', 'label': 'motorcycle'}, {'video_name': 'uncle_60450', 'label': 'uncle'}, {'video_name': 'a_00295', 'label': 'a'}, {'video_name': 'autumn_04288', 'label': 'autumn'}, {'video_name': 'last_66010', 'label': 'last'}, {'video_name': 'egg_18485', 'label': 'egg'}, {'video_name': 'worry_63872', 'label': 'worry'}, {'video_name': 'museum_37332', 'label': 'museum'}, {'video_name': 'comb_11812', 'label': 'comb'}, {'video_name': 'college_11704', 'label': 'college'}, {'video_name': 'son_53258', 'label': 'son'}, {'video_name': 'vegetable_61364', 'label': 'vegetable'}, {'video_name': 'morning_36856', 'label': 'morning'}, {'video_name': 'hate_26697', 'label': 'hate'}, {'video_name': 'patient_66270', 'label': 'patient'}, {'video_name': 'camera_08782', 'label': 'camera'}, {'video_name': 'hungry_28375', 'label': 'hungry'}, {'video_name': 'medicine_66111', 'label': 'medicine'}, {'video_name': 'r_45829', 'label': 'r'}, {'video_name': 'wear_62607', 'label': 'wear'}, {'video_name': 'to_58629', 'label': 'to'}, {'video_name': 'saturday_49327', 'label': 'saturday'}, {'video_name': 'principal_66331', 'label': 'principal'}, {'video_name': 'work_63769', 'label': 'work'}, {'video_name': 'drama_17557', 'label': 'drama'}, {'video_name': 'decorate_15092', 'label': 'decorate'}, {'video_name': 'march_34861', 'label': 'march'}, {'video_name': 'house_28157', 'label': 'house'}, {'video_name': 'good_25037', 'label': 'good'}, {'video_name': 'make_34558', 'label': 'make'}, {'video_name': 'policeman_43536', 'label': 'policeman'}, {'video_name': 'eraser_19512', 'label': 'eraser'}, {'video_name': 'time_58488', 'label': 'time'}, {'video_name': 'classroom_11006', 'label': 'classroom'}, {'video_name': 'with_63574', 'label': 'with'}, {'video_name': 'place_42983', 'label': 'place'}, {'video_name': 'wrong_64082', 'label': 'wrong'}, {'video_name': 'not_38765', 'label': 'not'}, {'video_name': 'near_37773', 'label': 'near'}, {'video_name': 'respect_47602', 'label': 'respect'}, {'video_name': 'poverty_44018', 'label': 'poverty'}, {'video_name': 'color_11752', 'label': 'color'}, {'video_name': 'give_69343', 'label': 'give'}, {'video_name': 'bad_04694', 'label': 'bad'}, {'video_name': 'very_69523', 'label': 'very'}, {'video_name': 'phone_42419', 'label': 'phone'}, {'video_name': 'thirsty_58023', 'label': 'thirsty'}, {'video_name': 'sin_51771', 'label': 'sin'}, {'video_name': 'appreciate_03067', 'label': 'appreciate'}, {'video_name': 'concentrate_12387', 'label': 'concentrate'}, {'video_name': 'chat_10099', 'label': 'chat'}, {'video_name': 'tea_57034', 'label': 'tea'}, {'video_name': 'world_63830', 'label': 'world'}, {'video_name': 'carrot_69259', 'label': 'carrot'}, {'video_name': 'india_29399', 'label': 'india'}, {'video_name': 'principle_44548', 'label': 'principle'}, {'video_name': 'neighbor_37945', 'label': 'neighbor'}, {'video_name': 'orange_40114', 'label': 'orange'}, {'video_name': 'girl_24590', 'label': 'girl'}, {'video_name': 'right_68142', 'label': 'right'}, {'video_name': 'class_10965', 'label': 'class'}, {'video_name': 'april_03150', 'label': 'april'}, {'video_name': 'h_26074', 'label': 'h'}, {'video_name': 'danger_14644', 'label': 'danger'}, {'video_name': 'where_63079', 'label': 'where'}, {'video_name': 'cat_09431', 'label': 'cat'}, {'video_name': 'silver_51725', 'label': 'silver'}, {'video_name': 'start_54454', 'label': 'start'}, {'video_name': 'that_57669', 'label': 'that'}, {'video_name': 'milk_36044', 'label': 'milk'}, {'video_name': 'train_59298', 'label': 'train'}, {'video_name': 'on_39672', 'label': 'on'}, {'video_name': 'brown_07957', 'label': 'brown'}, {'video_name': 'bank_04971', 'label': 'bank'}, {'video_name': 'tired_58588', 'label': 'tired'}, {'video_name': 'iran_30559', 'label': 'iran'}, {'video_name': 'february_21355', 'label': 'february'}, {'video_name': 'duck_17945', 'label': 'duck'}, {'video_name': 'when_63062', 'label': 'when'}, {'video_name': 'shirt_51054', 'label': 'shirt'}, {'video_name': 'calm_08739', 'label': 'calm'}, {'video_name': 'a_lot_02123', 'label': 'a lot'}, {'video_name': 'banana_04895', 'label': 'banana'}, {'video_name': 'really_46339', 'label': 'really'}, {'video_name': 'grateful_25521', 'label': 'grateful'}, {'video_name': 'white_63191', 'label': 'white'}, {'video_name': 'purple_45432', 'label': 'purple'}, {'video_name': 'camel_08772', 'label': 'camel'}, {'video_name': 'few_21548', 'label': 'few'}, {'video_name': 'sugar_55656', 'label': 'sugar'}, {'video_name': 'damage_14601', 'label': 'damage'}, {'video_name': 'apple_02997', 'label': 'apple'}, {'video_name': 'table_56552', 'label': 'table'}, {'video_name': 'strawberry_69492', 'label': 'strawberry'}, {'video_name': 'bookstore_07126', 'label': 'bookstore'}, {'video_name': 'go_24857', 'label': 'go'}, {'video_name': 'again_01457', 'label': 'again'}, {'video_name': 'spring_54116', 'label': 'spring'}, {'video_name': 'help_27194', 'label': 'help'}, {'video_name': 'family_20976', 'label': 'family'}, {'video_name': 'sleep_52314', 'label': 'sleep'}, {'video_name': 'rainbow_45966', 'label': 'rainbow'}, {'video_name': 'black_06455', 'label': 'black'}, {'video_name': 'turkey_59953', 'label': 'turkey'}, {'video_name': 'sweet_69499', 'label': 'sweet'}, {'video_name': 'close_11252', 'label': 'close'}, {'video_name': 'find_69323', 'label': 'find'}, {'video_name': 'january_30917', 'label': 'january'}, {'video_name': 'responsible_47645', 'label': 'responsible'}, {'video_name': 'blue_06822', 'label': 'blue'}, {'video_name': 'congratulations_12636', 'label': 'congratulations'}, {'video_name': 'want_62241', 'label': 'want'}, {'video_name': 'australia_04183', 'label': 'australia'}, {'video_name': 'tell_57273', 'label': 'tell'}, {'video_name': 'leave_32673', 'label': 'leave'}, {'video_name': 'understand_60543', 'label': 'understand'}, {'video_name': 'watermelon_62515', 'label': 'watermelon'}, {'video_name': 'restaurant_47655', 'label': 'restaurant'}, {'video_name': 'hear_26938', 'label': 'hear'}, {'video_name': 'will_63358', 'label': 'will'}, {'video_name': 'salt_49127', 'label': 'salt'}, {'video_name': 'man_34685', 'label': 'man'}, {'video_name': 'school_49577', 'label': 'school'}, {'video_name': 'emergency_18939', 'label': 'emergency'}, {'video_name': 'see_50107', 'label': 'see'}, {'video_name': 'please_43213', 'label': 'please'}, {'video_name': 'cow_13681', 'label': 'cow'}, {'video_name': 'airplane_01721', 'label': 'airplane'}, {'video_name': 'agree_01572', 'label': 'agree'}, {'video_name': 'outside_40463', 'label': 'outside'}, {'video_name': 'soup_53427', 'label': 'soup'}, {'video_name': 'purpose_45447', 'label': 'purpose'}, {'video_name': 'library_33048', 'label': 'library'}, {'video_name': 'conversation_13087', 'label': 'conversation'}, {'video_name': 'brother_07929', 'label': 'brother'}, {'video_name': 'calculator_08634', 'label': 'calculator'}, {'video_name': 'search_49931', 'label': 'search'}, {'video_name': 'pour_43999', 'label': 'pour'}, {'video_name': 'stadium_54248', 'label': 'stadium'}, {'video_name': 'noon_38614', 'label': 'noon'}, {'video_name': 'ice_cream_28581', 'label': 'ice cream'}, {'video_name': 'shame_50809', 'label': 'shame'}, {'video_name': 'city_10888', 'label': 'city'}, {'video_name': 'all_01912', 'label': 'all'}, {'video_name': 'yellow_64260', 'label': 'yellow'}, {'video_name': 'evening_19739', 'label': 'evening'}, {'video_name': 'road_48264', 'label': 'road'}, {'video_name': 'mine_36139', 'label': 'mine'}, {'video_name': 'dog_17076', 'label': 'dog'}, {'video_name': 'sunday_55768', 'label': 'sunday'}, {'video_name': 'difficult_16211', 'label': 'difficult'}, {'video_name': 'first_22071', 'label': 'first'}, {'video_name': 'enjoy_19255', 'label': 'enjoy'}, {'video_name': 'now_38982', 'label': 'now'}, {'video_name': 'in_29171', 'label': 'in'}, {'video_name': 'me_35267', 'label': 'me'}, {'video_name': 'have_26757', 'label': 'have'}, {'video_name': 'bell_05869', 'label': 'bell'}, {'video_name': 'hurry_28423', 'label': 'hurry'}, {'video_name': 'pencil_41819', 'label': 'pencil'}, {'video_name': 'boy_07436', 'label': 'boy'}, {'video_name': 'food_22739', 'label': 'food'}, {'video_name': 'pink_42827', 'label': 'pink'}, {'video_name': 'weather_62637', 'label': 'weather'}, {'video_name': 'sea_69459', 'label': 'sea'}, {'video_name': 'husband_28461', 'label': 'husband'}, {'video_name': 'stop_54915', 'label': 'stop'}, {'video_name': 'true_59783', 'label': 'true'}, {'video_name': 'cake_08608', 'label': 'cake'}, {'video_name': 'sorry_53363', 'label': 'sorry'}, {'video_name': 'moon_36774', 'label': 'moon'}, {'video_name': 'sun_55810', 'label': 'sun'}, {'video_name': 'people_41920', 'label': 'people'}, {'video_name': 'mother_36927', 'label': 'mother'}, {'video_name': 'before_05724', 'label': 'before'}, {'video_name': 'june_31345', 'label': 'june'}, {'video_name': 'book_69241', 'label': 'book'}, {'video_name': 'come_11839', 'label': 'come'}, {'video_name': 'repeat_47295', 'label': 'repeat'}, {'video_name': 'car_69258', 'label': 'car'}, {'video_name': 'you_64351', 'label': 'you'}, {'video_name': 'hurt_28438', 'label': 'hurt'}, {'video_name': 'dream_17624', 'label': 'dream'}, {'video_name': 'relax_47058', 'label': 'relax'}, {'video_name': 'trouble_59734', 'label': 'trouble'}, {'video_name': 'room_48507', 'label': 'room'}, {'video_name': 'pride_44485', 'label': 'pride'}, {'video_name': 'beautiful_05554', 'label': 'beautiful'}, {'video_name': 'habit_26079', 'label': 'habit'}, {'video_name': 'doctor_17007', 'label': 'doctor'}, {'video_name': 'we_62583', 'label': 'we'}, {'video_name': 'who_63219', 'label': 'who'}, {'video_name': 'how_28187', 'label': 'how'}, {'video_name': 'penalty_41810', 'label': 'penalty'}, {'video_name': 'day_14780', 'label': 'day'}, {'video_name': 'meet_35506', 'label': 'meet'}, {'video_name': 'plant_43078', 'label': 'plant'}, {'video_name': 'country_13542', 'label': 'country'}, {'video_name': 'july_31297', 'label': 'july'}, {'video_name': 'necessary_37827', 'label': 'necessary'}, {'video_name': 'truth_59839', 'label': 'truth'}, {'video_name': 'olympics_39646', 'label': 'olympics'}, {'video_name': 'bus_08330', 'label': 'bus'}, {'video_name': 'name_37575', 'label': 'name'}, {'video_name': 'far_21052', 'label': 'far'}, {'video_name': 'our_40340', 'label': 'our'}, {'video_name': 'discipline_16502', 'label': 'discipline'}, {'video_name': 'hour_28134', 'label': 'hour'}, {'video_name': 'movie_37120', 'label': 'movie'}, {'video_name': 'lawyer_32473', 'label': 'lawyer'}, {'video_name': 'year_64201', 'label': 'year'}, {'video_name': 'preach_44137', 'label': 'preach'}, {'video_name': 'heart_27013', 'label': 'heart'}, {'video_name': 'thursday_58359', 'label': 'thursday'}, {'video_name': 'include_29275', 'label': 'include'}, {'video_name': 'onion_39846', 'label': 'onion'}, {'video_name': 'forgive_22973', 'label': 'forgive'}, {'video_name': 'something_53228', 'label': 'something'}, {'video_name': 'student_55332', 'label': 'student'}, {'video_name': 'farmer_21085', 'label': 'farmer'}, {'video_name': 'transform_59379', 'label': 'transform'}, {'video_name': 'hot_28074', 'label': 'hot'}, {'video_name': 'bottle_07284', 'label': 'bottle'}, {'video_name': 'lion_33390', 'label': 'lion'}, {'video_name': 'japan_30927', 'label': 'japan'}, {'video_name': 'next_38222', 'label': 'next'}, {'video_name': 'office_39454', 'label': 'office'}, {'video_name': 'government_25240', 'label': 'government'}, {'video_name': 'safe_48995', 'label': 'safe'}, {'video_name': 'here_27254', 'label': 'here'}, {'video_name': 'what_62944', 'label': 'what'}, {'video_name': 'joke_31177', 'label': 'joke'}, {'video_name': 'play_69433', 'label': 'play'}, {'video_name': 'address_01011', 'label': 'address'}, {'video_name': 'like_69389', 'label': 'like'}, {'video_name': 'complain_12188', 'label': 'complain'}, {'video_name': 'chair_09847', 'label': 'chair'}, {'video_name': 'angry_02564', 'label': 'angry'}, {'video_name': 'china_10514', 'label': 'china'}, {'video_name': 'no_38482', 'label': 'no'}, {'video_name': 'red_46712', 'label': 'red'}, {'video_name': 'wife_63325', 'label': 'wife'}, {'video_name': 'sister_51882', 'label': 'sister'}, {'video_name': 'england_19218', 'label': 'england'}, {'video_name': 'new_38103', 'label': 'new'}, {'video_name': 'number_39031', 'label': 'number'}, {'video_name': 'afraid_01372', 'label': 'afraid'}, {'video_name': 'grey_68502', 'label': 'grey'}, {'video_name': 'fun_23799', 'label': 'fun'}, {'video_name': 'trust_59827', 'label': 'trust'}, {'video_name': 'friend_23567', 'label': 'friend'}, {'video_name': 'think_57954', 'label': 'think'}, {'video_name': 'complete_12217', 'label': 'complete'}, {'video_name': 'from_23637', 'label': 'from'}, {'video_name': 'fish_22109', 'label': 'fish'}, {'video_name': 'speak_53611', 'label': 'speak'}, {'video_name': 'sky_69477', 'label': 'sky'}, {'video_name': 'december_15017', 'label': 'december'}, {'video_name': 'today_58676', 'label': 'today'}, {'video_name': 'bring_07803', 'label': 'bring'}
    ]
    augmented_dataset = augment_dataset(landmark_dir, dataset, num_augmentations=4)
    # Save augmented dataset
    pd.DataFrame(augmented_dataset).to_csv(
        f"{landmark_dir}/augmented_dataset.csv", index=False
    )
````

````python
import torch
from torch.utils.data import Dataset, DataLoader
import pandas as pd
import numpy as np
import pickle
class SignDataset(Dataset):
    def __init__(self, df, landmark_dir):
        self.df = df
        self.landmark_dir = landmark_dir
        self.label_to_idx = {label: idx for idx, label in enumerate(sorted(df['label'].unique()))}
    def __len__(self):
        return len(self.df)
    def __getitem__(self, idx):
        entry = self.df.iloc[idx]
        with open(f"{self.landmark_dir}/{entry['video_name']}/landmarks_preprocessed.pkl", 'rb') as f:
            landmarks = pickle.load(f)
        # Convert dict to array if needed
        if isinstance(landmarks, dict):
            landmarks = np.concatenate([
                np.array(landmarks.get('face', np.zeros((30, 468, 3)))),
                np.array(landmarks.get('left_hand', np.zeros((30, 21, 3)))),
                np.array(landmarks.get('right_hand', np.zeros((30, 21, 3)))),
                np.array(landmarks.get('pose', np.zeros((30, 33, 3))))
            ], axis=1)  # [frames, 543, 3]
        # Normalize
        landmarks = (landmarks - landmarks.mean(axis=(0, 1))) / (landmarks.std(axis=(0, 1)) + 1e-6)
        # Pad/truncate to 30 frames
        if landmarks.shape[0] < 30:
            landmarks = np.pad(landmarks, ((0, 30 - landmarks.shape[0]), (0, 0), (0, 0)), mode='edge')
        elif landmarks.shape[0] > 30:
            landmarks = landmarks[:30]
        return torch.tensor(landmarks, dtype=torch.float32), self.label_to_idx[entry['label']]
# Load dataset
landmark_dir = '/content/drive/MyDrive/Datasets_ISL-CSLRT-PSL/Processed_Landmarks_WLASL'
df = pd.read_csv(f"{landmark_dir}/augmented_dataset.csv")
train_df = df.sample(frac=0.8, random_state=42)
val_df = df.drop(train_df.index)
train_dataset = SignDataset(train_df, landmark_dir)
val_dataset = SignDataset(val_df, landmark_dir)
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=16)
````

````python
# Check Frames before model building
import os
import pickle
import pandas as pd
import numpy as np
from pathlib import Path
from collections import Counter

def check_frame_size(landmark_dir, dataset_csv):
    """
    Check the frame size of landmarks_preprocessed.pkl files in the dataset.
    Args:
        landmark_dir: Path to landmark directory (e.g., '/content/drive/.../Processed_Landmarks_WLASL').
        dataset_csv: Path to augmented_dataset.csv.
    Returns:
        Frame size statistics and distribution.
    """
    landmark_dir = Path(landmark_dir)
    df = pd.read_csv(dataset_csv)
    frame_sizes = []

    for idx, row in df.iterrows():
        video_name = row['video_name']
        landmark_path = landmark_dir / video_name / 'landmarks_preprocessed.pkl'

        if not landmark_path.exists():
            print(f"Warning: {landmark_path} does not exist, skipping")
            continue

        # Load landmarks
        with open(landmark_path, 'rb') as f:
            landmarks = pickle.load(f)

        # Extract frame count
        if isinstance(landmarks, dict):
            # Assume dictionary with face, hands, pose; use face as reference
            frames = np.array(landmarks.get('face', np.zeros((1, 468, 3)))).shape[0]
        elif isinstance(landmarks, np.ndarray):
            frames = landmarks.shape[0]
        else:
            print(f"Warning: Unexpected format in {video_name}, skipping")
            continue

        frame_sizes.append(frames)

    # Compute statistics
    if frame_sizes:
        frame_sizes = np.array(frame_sizes)
        stats = {
            'total_videos': len(frame_sizes),
            'mean_frames': frame_sizes.mean(),
            'min_frames': frame_sizes.min(),
            'max_frames': frame_sizes.max(),
            'median_frames': np.median(frame_sizes),
            'std_frames': frame_sizes.std()
        }
        # Frame size distribution
        distribution = Counter(frame_sizes)

        print("\nFrame Size Statistics:")
        for key, value in stats.items():
            print(f"  {key}: {value:.2f}")
        print("\nFrame Size Distribution (Top 5):")
        for frames, count in sorted(distribution.items(), key=lambda x: x[1], reverse=True)[:5]:
            print(f"  {frames} frames: {count} videos")

        return frame_stats, distribution
    else:
        print("Error: No valid landmark files found")
        return None, None

if __name__ == "__main__":
    landmark_dir = '/content/drive/MyDrive/Datasets_ISL-CSLRT-PSL/Processed_Landmarks_WLASL'
    dataset_csv = f"{landmark_dir}/augmented_dataset.csv"
    stats, distribution = check_frame_size(landmark_dir, dataset_csv)
````

````python
import torch
import torch.nn as nn
import pandas as pd
import numpy as np
import pickle
from torch.utils.data import Dataset, DataLoader
from pathlib import Path
import logging
import os

# Setup logging
logging.basicConfig(filename='training_log.txt', level=logging.INFO,
                    format='%(asctime)s - %(message)s')

class SignCNNLSTM(nn.Module):
    def __init__(self, num_keypoints=543, num_classes=396, num_frames=190):
        """
        SignCNNLSTM for WLASL sign recognition with 190 frames.
        Args:
            num_keypoints: Number of keypoints (e.g., 543 for MediaPipe).
            num_classes: Number of glosses (~396).
            num_frames: Fixed frame length (190).
        """
        super().__init__()
        self.cnn = nn.Sequential(
            # Input: [batch, 3, 190, 543, 1]
            nn.Conv3d(3, 64, kernel_size=(3, 3, 1), padding=(1, 1, 0)),
            nn.ReLU(),
            nn.MaxPool3d((2, 2, 1)),  # [batch, 64, 95, 271, 1]
            nn.Conv3d(64, 128, kernel_size=(3, 3, 1), padding=(1, 1, 0)),
            nn.ReLU(),
            nn.MaxPool3d((2, 2, 1)),  # [batch, 128, 47, 135, 1]
            nn.Conv3d(128, 256, kernel_size=(3, 3, 1), padding=(1, 1, 0)),
            nn.ReLU(),
            nn.MaxPool3d((2, 2, 1)),  # [batch, 256, 23, 67, 1]
            nn.Flatten(start_dim=2)   # [batch, 256, 23*67=1541]
        )
        self.lstm = nn.LSTM(256, 512, num_layers=2, batch_first=True, dropout=0.3)
        self.fc = nn.Sequential(
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        # x: [batch, 190, 543, 3] → [batch, 3, 190, 543, 1]
        x = x.permute(0, 3, 1, 2).unsqueeze(-1)
        x = self.cnn(x)  # [batch, 256, 1541]
        x = x.permute(0, 2, 1)  # [batch, 1541, 256]
        x, _ = self.lstm(x)  # [batch, 1541, 512]
        return self.fc(x[:, -1, :])  # [batch, num_classes]

class WLASLDataset(Dataset):
    def __init__(self, df, landmark_dir, num_frames=190):
        """
        Dataset for loading WLASL landmark data.
        Args:
            df: DataFrame from augmented_dataset.csv.
            landmark_dir: Path to landmark directory.
            num_frames: Fixed frame length (190).
        """
        self.df = df
        self.landmark_dir = landmark_dir
        self.num_frames = num_frames
        self.label_to_idx = {label: idx for idx, label in enumerate(sorted(df['label'].unique()))}

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        entry = self.df.iloc[idx]
        landmark_path = Path(self.landmark_dir) / entry['video_name'] / 'landmarks_preprocessed.pkl'
        try:
            with open(landmark_path, 'rb') as f:
                landmarks = pickle.load(f)
        except FileNotFoundError:
            logging.warning(f"Landmark file not found: {landmark_path}")
            return torch.zeros(self.num_frames, 543, 3), self.label_to_idx[entry['label']]

        # Convert dict to array if needed
        if isinstance(landmarks, dict):
            landmarks = np.concatenate([
                np.array(landmarks.get('face', np.zeros((190, 468, 3)))),
                np.array(landmarks.get('left_hand', np.zeros((190, 21, 3)))),
                np.array(landmarks.get('right_hand', np.zeros((190, 21, 3)))),
                np.array(landmarks.get('pose', np.zeros((190, 33, 3))))
            ], axis=1)  # [190, 543, 3]

        # Normalize
        landmarks = (landmarks - landmarks.mean(axis=(0, 1))) / (landmarks.std(axis=(0, 1)) + 1e-6)

        # Ensure 190 frames
        frames = landmarks.shape[0]
        if frames != self.num_frames:
            logging.warning(f"Unexpected frame count {frames} in {entry['video_name']}")
            if frames < self.num_frames:
                landmarks = np.pad(landmarks, ((0, self.num_frames - frames), (0, 0), (0, 0)), mode='edge')
            else:
                landmarks = landmarks[:self.num_frames]

        return torch.tensor(landmarks, dtype=torch.float32), self.label_to_idx[entry['label']]

def train_wlasl_model(model, train_loader, val_loader, num_epochs=50, lr=0.001, device='cuda', save_path='model'):
    """
    Train the SignCNNLSTM model on WLASL data.
    """
    model = model.to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=5, factor=0.5)

    metrics = []
    best_val_acc = 0
    os.makedirs(os.path.dirname(save_path), exist_ok=True)

    for epoch in range(num_epochs):
        model.train()
        train_loss = 0
        train_correct = 0
        train_total = 0
        for landmarks, labels in train_loader:
            landmarks, labels = landmarks.to(device), labels.to(device)
            outputs = model(landmarks)
            loss = criterion(outputs, labels)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            train_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            train_total += labels.size(0)
            train_correct += (predicted == labels).sum().item()

        model.eval()
        val_loss = 0
        val_correct = 0
        val_total = 0
        with torch.no_grad():
            for landmarks, labels in val_loader:
                landmarks, labels = landmarks.to(device), labels.to(device)
                outputs = model(landmarks)
                loss = criterion(outputs, labels)
                val_loss += loss.item()
                _, predicted = torch.max(outputs, 1)
                val_total += labels.size(0)
                val_correct += (predicted == labels).sum().item()

        train_loss /= len(train_loader)
        train_acc = 100 * train_correct / train_total
        val_loss /= len(val_loader)
        val_acc = 100 * val_correct / val_total
        scheduler.step(val_loss)

        # Log metrics
        log_msg = (f"Epoch {epoch+1}/{num_epochs}, Train Loss: {train_loss:.4f}, "
                   f"Train Acc: {train_acc:.2f}%, Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.2f}%")
        print(log_msg)
        logging.info(log_msg)
        metrics.append([epoch+1, train_loss, train_acc, val_loss, val_acc])

        # Save best model
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            torch.save(model.state_dict(), save_path)
            logging.info(f"Saved best model at epoch {epoch+1} with Val Acc: {val_acc:.2f}%")

    # Save metrics
    pd.DataFrame(metrics, columns=['Epoch', 'Train Loss', 'Train Acc', 'Val Loss', 'Val Acc']).to_csv(
        f"{os.path.dirname(save_path)}/training_metrics.csv", index=False
    )

    return model

if __name__ == "__main__":
    # Paths
    landmark_dir = '/content/drive/MyDrive/Datasets_ISL-CSLRT-PSL/Processed_Landmarks_WLASL'
    dataset_csv = f"{landmark_dir}/augmented_dataset.csv"
    model_save_path = f"{landmark_dir}/wlasl_sign_cnn_lstm_190frames.pth"

    # Load dataset
    df = pd.read_csv(dataset_csv)
    print(f"Dataset size: {len(df)}, Classes: {df['label'].nunique()}")
    train_df = df.sample(frac=0.8, random_state=42)
    val_df = df.drop(train_df.index)

    # Create datasets
    train_dataset = WLASLDataset(train_df, landmark_dir, num_frames=190)
    val_dataset = WLASLDataset(val_df, landmark_dir, num_frames=190)
    train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True, num_workers=2, pin_memory=True)
    val_loader = DataLoader(val_dataset, batch_size=16, num_workers=2, pin_memory=True)

    # Initialize model
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = SignCNNLSTM(num_keypoints=543, num_classes=df['label'].nunique(), num_frames=190)

    # Train
    trained_model = train_wlasl_model(model, train_loader, val_loader, num_epochs=50, lr=0.001,
                                     device=device, save_path=model_save_path)
````
